```python
import sqlite3
import csv

def create_database():
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS books (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            title TEXT,
            author TEXT,
            genre TEXT,
            publication_date TEXT,
            isbn TEXT
        )
    ''')

    conn.commit()
    conn.close()

def add_book():
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    title = input("Enter book title: ")
    author = input("Enter author name: ")
    genre = input("Enter genre: ")
    publication_date = input("Enter publication date (YYYY-MM-DD): ")
    isbn = input("Enter ISBN: ")

    cursor.execute('''
        INSERT INTO books (title, author, genre, publication_date, isbn)
        VALUES (?, ?, ?, ?, ?)
    ''', (title, author, genre, publication_date, isbn))

    conn.commit()
    conn.close()

def filter_books():
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    filter_by = input("Filter by (title, author, genre, publication_date, isbn): ")
    filter_value = input("Enter filter value: ")

    cursor.execute(f'''
        SELECT * FROM books WHERE {filter_by} LIKE ?
    ''', (f"%{filter_value}%",))

    rows = cursor.fetchall()
    for row in rows:
        print(row)

    conn.close()

def export_data():
    conn = sqlite3.connect('books.db')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM books")
    rows = cursor.fetchall()

    with open('books.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['ID', 'Title', 'Author', 'Genre', 'Publication Date', 'ISBN'])
        writer.writerows(rows)

    conn.close()

def main():
    create_database()

    while True:
        print("\nInventory Management System")
        print("1. Add a new book")
        print("2. Filter books")
        print("3. Export book data")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            add_book()
        elif choice == '2':
            filter_books()
        elif choice == '3':
            export_data()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
